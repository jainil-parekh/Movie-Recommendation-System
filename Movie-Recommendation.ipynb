```python
#ast is used to evaluate a string conntaining python literal. In this it used to obtain list from the string of list
#i.e. genres column is of type'[{}]' and we wanted it in the form of []
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#converting string into list so that we can CONCAT
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
#removing the space between words, in order to make the search more efficient
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "")for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

CountVectorizer:-
CountVectorizer is a tool for converting a collection of text documents into a matrix of token counts. It has various parameters that allow customization of the tokenization and feature extraction process, such as handling stop words, defining n-gram ranges, and setting minimum or maximum document frequency thresholds.

fit_transform:-
fit_transform is a method available on CountVectorizer and other transformers in scikit-learn. It performs two main tasks in one step:

Fit: Learns the vocabulary dictionary from the training data (i.e., identifies all unique tokens/words in the text data and their frequencies).
Transform: Transforms the training data into a document-term matrix using the learned vocabulary.

```python
#this countvectorizer :- Converts the text documents into a sparse matrix where each row represents a document and each column represents a token from the vocabulary
#stop_words means, words like are, am, is etc. will not be considered at the time of analysis.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
#fit_transform :- conversion of raw text data into numerical vectors
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#This will find similarity based on angle and not distance, smaller the angle more the similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]#used to find index
    distances = similarity[movie_index] #finding the similar ones and storing it in distances
    movie_list = sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
